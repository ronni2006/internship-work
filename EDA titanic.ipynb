
# Outlier Detection
1. Boxplot (Visual Method)
Best for: Initial exploration of numerical data.

2. IQR (Interquartile Range) Method
The Interquartile Range (IQR) measures the spread of the middle 50% of the data.
IQR=Q3-Q1
Calculate the limits:
• Lower Limit = Q1 - 1.5 × IQR
• Upper Limit = Q3 + 1.5 x IQR
Best for: Skewed or non-normal data distributions.

3. Z-Score Method
The Z-score measures how many standard deviations a data point is away from the mean.
Z=X-Mean/Standard Deviation
If IZI>3 the observation is generally considered an outlier.

Best for: Normally distributed data.

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size= 0.2, random_state= 42)
print(x_train, x_test, y_train, y_test)
```

# Logistic Regression

Definition

Despite its name, logistic regression is a classification algorithm. It estimates the probability that an input belongs to a particular class using 
the logistic (sigmoid function, outputting values between 0 and 1.

How It Works

Applies the sigmoid function o(z) = 1/ (1 + ez) to transform a
linear combination of inputs into a probability. A threshold (typically 0.5) determines the final class label.

-

# Linear Combination:

Linear Combination is the first step in Logistic Regression. It means the model combines all the input features using their respective weights and adds a bias to produce a single value (z).

Formula

z=w1x1+w2x2+w3x3+⋯+wnxn+b

Where:

x₁, x₂, …, xₙ = Input features

w₁, w₂, …, wₙ = Weights learned by the model

b = Bias (intercept)

z = Linear combination (a single numeric value)

# sigmoid:
use when output is in two variables

# Probability:

Probability inLogistic Regression

Probability is the likelihood that a data point belongs to a particular class. In Logistic Regression, after calculating the linear combination (z), the model uses the sigmoid function to convert it into a probability value between 0 and 1.

Formula
P(y=1)=1/1+e−z1

# Class Label:

A Class Label is the final prediction made by the Logistic Regression model after converting the probability into a category.

The model first calculates a probability (between 0 and 1) and then compares it with a threshold (usually 0.5):

If Probability ≥ 0.5 → Class Label = 1 (Positive Class)

If Probability < 0.5 → Class Label = 0 (Negative Class)

Formula:

Class Label={1,0,if P≥0.5  if P<0.5

Where:

P = Probability predicted by the sigmoid function.

Definition:

SVM finds the optimal hyperplane that separates classes with the
maximum margin. Data points closest to the boundary are called support vectors - they define the decision surface.

How It Works:

Maximizes the margin between classes.

For non-linearly separable data, SVM uses the kernel trick (RBF, polynomial) to map inputs into higher-dimensional space where separation is possible.






Naive Bayes

Definition

A probabilistic classifier based on Bayes' Theorem with a naive" assumption: all features are conditionally independent given the class label. Surprisingly effective despite this simplification.

How It Works

Calculates P(Class | Features) using prior probabilities and likelihoods. The class with the highest posterior probability is selected. Extremely fast to train and predict.

Real-World Applications

Spam filtering, sentiment analysis, document classification, news categorization, and medical screening tools.

Task Type

Classification- especially effective for high-dimensional text data and real-time applications.
