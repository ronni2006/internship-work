# Scaling Methods in Machine Learning:

Standard Scaler vs MinMax Scaler vs Robust Scaler

1. Standard Scaler

Formula:

Key Points:

2. MinMax Scaler

Formula:

(X−Xmin)


Key Points:

Range: [0, 1]
Highly sensitive to outliers
Compresses normal values
3. Robust Scaler

Formula:

X
scale=IQR(X−median)

Where:

Median = Middle value of the data
IQR (Interquartile Range) = Q3 − Q1

Key Points:

Based on median & IQR
Robust to outliers
Best for skewed data
Original Data with Outliers

The image illustrates a dataset containing mostly normal values along with a few extreme outliers. It then shows how each scaling method transforms the data:

Standard Scaler: Centers the data around 0 with standard deviation 1, but outliers still influence the scaling.
MinMax Scaler: Scales all values to the range [0, 1], causing normal values to become compressed when outliers exist.
Robust Scaler: Uses the median and IQR, making it much less affected by outliers and preserving the spread of the majority of the data.
